```python
import random
word_list = ["aardvark", "baboon", "camel"]
rand=random.randint(0,2)
chosen_word=word_list[rand]
```

```python
stages = ['''
 +---+
 | |
 O |
/|\ |
/ \ |
 |
=========
''', '''
 +---+
 | |
 O |
/|\ |
/ |
 |
=========
''', '''
 +---+
 | |
 O |
/|\ |
 |
 |
=========
''', '''
 +---+
 | |
 O |
/| |
 |
 |
=========''', '''
 +---+
 | |
 O |
 | |
 |
 |
=========
''', '''
 +---+
 | |
 O |
 |
 |
 |
=========
''', '''
 +---+
 | |
 |
 |
 |
 |
=========
''']
```

```python
# input_user=(input("Guess any letter"))
# guess=input_user.lower()
display=["_"]*len(chosen_word)
lives=6
string=' '.join(display)
while "_" in display and lives>0:
        
        print(stages[lives])

        display_string=' '.join(display)
        print(display_string)

        guess=(input("Guess any letter")).lower()

        if guess in display:
            print(f"You've already guessed '{guess}'. Try another letter.")
            continue
        if not guess.isalpha() or len(guess) != 1:
            print("Please enter a single letter.")
            continue
        found = False
        for i in range(len(chosen_word)):
          if chosen_word[i]==guess:
             display[i]=guess
             found = True
             

        if not found:
            lives -= 1
            print(f"'{guess}' is not in the word. You lose a life!")
            if lives == 0:
                  print("You lose.")
                  break
print("Congratulations! You've guessed the word:", ''.join(display))
        
```
